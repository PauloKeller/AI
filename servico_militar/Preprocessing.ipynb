## OUT_OF_YEAR

Caso esteja fora do ano de alistamento = 1

```python
def calculate_out_of_year(born_year, enlistment_year):
    age = enlistment_year - born_year
    return age >= 19
```

```python
out_of_year = calculate_out_of_year(pre_uf_region_csv['ANO_NASCIMENTO'], pre_uf_region_csv['VINCULACAO_ANO'])
pre_uf_region_csv['OUT_OF_YEAR'] = out_of_year
pre_uf_region_csv['OUT_OF_YEAR'] = np.where(pre_uf_region_csv['OUT_OF_YEAR'], 1, 0)
pre_uf_region_csv = pre_uf_region_csv.drop(['ANO_NASCIMENTO', 'VINCULACAO_ANO'], axis=1)
```

```python
pre_out_of_year_csv = pre_uf_region_csv.copy()
pre_out_of_year_csv.head()
```
