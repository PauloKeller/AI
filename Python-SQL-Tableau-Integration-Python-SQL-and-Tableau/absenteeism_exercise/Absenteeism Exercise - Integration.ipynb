<h2>Creating the INSERT Statement</h2>

```python
df_new_obs['Age']
```

```python
insert_query = 'INSERT INTO predicted_outputs VALUES '
insert_values = []

for i in range(df_new_obs.shape[0]): 
    pred_value = ""
    for j in range(df_new_obs.shape[1]):
        if (j >= df_new_obs.shape[1]-1):
            pred_value += str(df_new_obs[df_new_obs.columns.values[j]][i])
        else:
            pred_value += str(df_new_obs[df_new_obs.columns.values[j]][i]) + ','
    insert_values.append('(' + pred_value + ')')
    
for i in insert_values:
    insert_query += i + ','

insert_query = insert_query[:-1] + ';'
```

```python
insert_query
```

```python
cursor.execute(insert_query)

conn.commit()
```
