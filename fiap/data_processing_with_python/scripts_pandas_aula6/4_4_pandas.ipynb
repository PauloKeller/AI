### Agrupamento

```python
# Exemplo de dados
dados = pd.DataFrame({'grupo': ['A', 'A', 'B', 'B', 'A'],
                      'valor': [1, 2, 3, 4, 5],
                      'quant': [10, 20, 30, 40, 50]})

# Agrupando os dados por 'grupo' e calculando a mÃ©dia dos valores para cada grupo
media_por_grupo = dados.groupby('grupo')['quant'].mean()

print(media_por_grupo)
```

### Apply

```python
# Exemplo de dados
dados = pd.DataFrame({'valores': [1, 2, 3, 4, 5]})

# FunÃ§Ã£o para calcular o quadrado de um nÃºmero
def calcular_quadrado(x):
    return x ** 2

# Aplicando a funÃ§Ã£o 'calcular_quadrado' Ã  coluna 'valores' usando apply
dados['quadrado'] = dados['valores'].apply(calcular_quadrado)

print(dados)
```

### Merge

```python
# DataFrame com informaÃ§Ãµes sobre os alunos
alunos = pd.DataFrame({
    'student_id': [1, 2, 3, 4],
    'nome': ['JoÃ£o', 'Maria', 'Pedro', 'Ana']
})

# DataFrame com informaÃ§Ãµes sobre as disciplinas
disciplinas = pd.DataFrame({
    'student_id': [1, 2, 2, 3, 4],
    'disciplina': ['MatemÃ¡tica', 'FÃ­sica', 'QuÃ­mica', 'HistÃ³ria', 'Geografia']
})

# Combinando os DataFrames usando merge
resultado = pd.merge(alunos, disciplinas, on='student_id')

print(resultado)
```

```python
# Left join

# DataFrame de exemplo para alunos
alunos = pd.DataFrame({
    'student_id': [1, 2, 3, 4],
    'nome': ['JoÃ£o', 'Maria', 'Pedro', 'Ana']
})

# DataFrame de exemplo para notas
notas = pd.DataFrame({
    'student_id': [1, 2, 3],
    'nota': [8, 7, 9]
})

# Left join entre alunos e notas
left_join = pd.merge(alunos, notas, on='student_id', how='left')

print("Left join:")
print(left_join)

# Right join entre alunos e notas
right_join = pd.merge(alunos, notas, on='student_id', how='right')

print("\nRight join:")
print(right_join)
```

### Query

```python
# Exemplo de dados
dados = pd.DataFrame({
    'idade': [25, 35, 45, 30, 40],
    'salÃ¡rio': [40000, 60000, 70000, 45000, 55000]
})

# Usando query para selecionar as linhas onde a idade Ã© maior que 30 e o salÃ¡rio Ã© maior que 50000
resultado = dados.query('idade > 30 and salÃ¡rio > 50000')

print(resultado)
```
